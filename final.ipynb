```python
# processing : DFi_2 (Dis-inv2(PilotData).txt)
import csv

delimiter_char = '\t'

expt_id = None
Group = None
group_id = None
participant_id = None
response_correct = None
response_name = None
response_rt = None
stimuli_presented = None

trial_template = None
trial_duration = None
PartBlocks = None
AllBlocks = None
Pattern = None
trial_num = None
PB = None
Final = None
Part = None
Correct = None
invRT = None
Trial = None  # trial count within the main experiment


t1 = ",expt_id,Group,group_id,invRT,Trial,participant_id,response_correct,response_name,response_rt,stimuli_presented,trial_template,trial_duration,PartBlocks,AllBlocks,Pattern,trial_num,PB,Final,Part,Correct"
temt1 = t1.split(",")
result = [temt1]

index = 1   # line number


with open('.ipynb_checkpoints/Dis-Inv2(PilotData).txt', mode='r', newline='') as file:
    csv_reader = csv.reader(file, delimiter=delimiter_char)

    for row in csv_reader:
        # Print row for debugging
        print(row)

        # ---------------------------
        # expt_id comes from Session ID
        # ---------------------------
        if len(row) == 2 and row[0] == '.Session ID':
            expt_id = row[1]

        # ---------------------------
        # Main trial rows
        # ---------------------------
        if len(row) > 4 and row[0] != "Participant" and row[0] != "Group":

            # Values we CAN extract safely
            Group = row[0]
            participant_id = row[1]
            trial_template = row[2] if len(row) > 2 else None
            stimuli_presented = row[5] if len(row) > 5 else None
            trial_duration = row[6] if len(row) > 6 else None
            response_correct = row[10] if len(row) > 10 else None
            Correct = response_correct  # same as above
            response_rt = row[12] if len(row) > 12 else None
            invRT = -1000 / float(response_rt) if response_rt else None
            PB = row[13] if len(row) > 13 else None

            # Values NOT FOUND  leave as None
            group_id = None
            response_name = None
            PartBlocks = None
            AllBlocks = None
            Pattern = None
            trial_num = None
            Final = None
            Part = None
            Trial = None

            # Append row to result
            result.append([
                index, expt_id, Group, group_id, invRT, Trial, participant_id,
                response_correct, response_name, response_rt, stimuli_presented,
                trial_template, trial_duration, PartBlocks, AllBlocks, Pattern,
                trial_num, PB, Final, Part, Correct
            ])
            index += 1

filename = 'dis-inv2.csv'

with open(filename, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(result)

print(f"Data successfully written to {filename}")

```
